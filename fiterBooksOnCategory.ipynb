```python

import os 
from helperFunctions import findISBN,getBookMetadata
import pprint
from collections import Counter
from tqdm.auto import tqdm
```

```python

def accumulateCategories():
    directory = r"/Users/humuyao/Downloads/Book3"
    folders = ['7', 'X', 'Y', 'Z']
#     folders = ['X']

    books_metadata = {}
    categories=[]
    booksWithFaultyISBN=0
    for folder in folders:
        subdir = directory + '/' + folder
        for filename in os.listdir(subdir):
            if filename.endswith(".txt"):
                filepath = subdir + '/' + filename
                text = open(filepath, "r",encoding='utf-8').read()

                print(filename)
                print(filepath)
                
                authorIncluded = True if "-" in filename else False
                title = filename if not authorIncluded else filename.split("-")[0].strip()
                author = False if not authorIncluded else filename.split("-")[1].split(".")[0].strip()
                ISBN = findISBN(text) # Books without ISBN have already been removed
                print(ISBN)
                
                if ISBN == False:
                    os.remove(filepath)
                    print('-'*100)
                    continue
                
                metadata = getBookMetadata(ISBN, title, author)

                # The ISBN does NOT match the Google ISBNs, remove book.
                if metadata == False:
                    booksWithFaultyISBN += 1
                    os.remove(filepath)
                    print('-'*100)
                    continue 
               

                books_metadata[title] = metadata
                categories+= metadata['categories']
                
#                 print(ISBN)

#                 if ISBN and len(ISBN) >=10:
#                     categories = getCategories(ISBN,filename) 
#                     print(categories)
#                 else:
#                     print('ERROR: No ISBN FOUND ')
#                 book_name = f"{title}-{author}"
#                 book_quotes = filterBook(filepath)
# #                 collection_idx = appendToIndex(collection_idx,book_name,book_quotes)
                
                print('-'*100)
                
    return books_metadata,categories
    

    
```

```python
books_metadata, categories = accumulateCategories()

```
